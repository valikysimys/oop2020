### Індивідуальне завдання №1
**Для кожного елементу об'єкту start додати 1. Виправити помилку в стартовому коді** 

```python
# стартовий код Індивідуального завдання №1
def addNum(start, num):
    for i in range(len(start)):
        start[i] += num
    return start
 
start = (1, 2, 3, 4)

end = addNum(start, 1)
 
print(start)
print(end)
```
